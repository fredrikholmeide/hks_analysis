```python
import pandas as pd
```

```python
df_results = pd.read_csv('Resultlist-23.csv', delimiter=';')
df_results.describe()
```

```python
df_results.head()
```

```python
# Select the columns that contain a number in the name
filtered_columns = [col for col in df_results.columns if (any(char.isdigit() for char in col) and col.endswith('_S') or col == 'Mål_S') and col != '1 Louisesgate_S']
```

```python
df_results[filtered_columns][0:30]
```

```python
def convert_to_seconds(time_str):
    # Check if the time is a string
    if isinstance(time_str, float):
        return time_str
    minutes, seconds = time_str.split(':')
    total_seconds = int(minutes) * 60 + int(seconds)
    return total_seconds
```

```python
filtered_columns
```

```python
# Convert the columns to numeric, the current format is mm:ss and we want to convert it to seconds
for col in filtered_columns:
    # Check if the column contains a number
        
        df_results[col] = df_results[col].apply(convert_to_seconds)

```

```python

# Remove all rows that contain zero
df_results = df_results[(df_results[filtered_columns] != 0).all(axis=1)]
df_results[filtered_columns]
```

```python
#Make a boxplot for each column
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
df_results[filtered_columns].boxplot()
plt.xticks(rotation=20)
plt.show()
```

```python
# Calculate the mean and standard deviation for each column
mean = df_results[filtered_columns].mean()
std = df_results[filtered_columns].std()
mean, std
```

```python
import seaborn as sns
plt.figure(figsize=(20,10))
plt.xticks(rotation=20)
sns.violinplot(data=df_results[filtered_columns])
```

```python
df_results[filtered_columns]
```

```python
# Remove the outliers, we will define an outlier as a value that is more than 3 standard deviations from the mean. 
# We will remove the entire row if any of the values are outliers
# The mean and standard deviation is calculated for each column, each column is treated independently
df_results = df_results[~((df_results[filtered_columns] - mean).abs() > 3 * std).any(axis=1)]
df_results[filtered_columns]
```

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=20)
sns.violinplot(data=df_results[filtered_columns])
```

```python
#Make a scatter plot with mean and standard deviation, and mark each column with a different color
plt.figure(figsize=(20,10))
plt.scatter(mean, std)
for i, txt in enumerate(filtered_columns):
    plt.annotate(txt, (mean[i], std[i]))
plt.xlabel('Mean')
plt.ylabel('Standard deviation')
plt.show()


```
